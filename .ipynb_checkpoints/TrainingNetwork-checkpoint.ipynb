# Imports

```python
import tensorflow as tf
import numpy as np 
from skimage import io
from skimage.io import imread
import matplotlib.pyplot as plt
import os
import glob
import pathlib
from sklearn.model_selection import train_test_split
from tensorflow.keras.layers import Conv2D, MaxPooling2D
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.models import load_model
from tensorflow.keras import models
from tensorflow.keras import layers
import seaborn as sns
```

# Loading Data

```python
DATADIR = "G:\Input_data2/"
CATEGORIES = ["class1","class2","Org"]
filenames=[]
file_path=[]
IMG_SIZE=200

for category in CATEGORIES:  
    clss_name, _ = os.path.splitext(category)          
    if clss_name == 'class1':
        path = os.path.join(DATADIR,clss_name) 
        for img in os.listdir(path): 
            name, _ = os.path.splitext(img)
            img_path = [path +"/"+name+".jpeg"]
            filenames = filenames+img_path
            img_array = io.imread(os.path.join(path,img))
            H = img_array.shape[0]
            W = img_array.shape[1]
            class_num = np.zeros((H,W))
            for i in range(H):
                for j in range(W):
                    if img_array.item(i,j) == 0:
                        class_num[i,j]=1.0
                    else:
                        class_num[i,j]=0.0
            file_path.append([img_path[0],class_num])     
            
    elif clss_name == 'Org':
        path = os.path.join(DATADIR,clss_name) 
        for img in os.listdir(path): 
            name, _ = os.path.splitext(img)
            img_path = [path +"/"+name+".jpeg"]
            filenames = filenames+img_path
            img_array = io.imread(os.path.join(path,img))
            H = img_array.shape[0]
            W = img_array.shape[1]
            class_num = np.zeros((H,W))
            for i in range(H):
                for j in range(W):
                    if img_array.item(i,j) == 0:
                        class_num[i,j]=1.0
                    else:
                        class_num[i,j]=0.35
            file_path.append([img_path[0],class_num])     
                 
    else:
        path = os.path.join(DATADIR,clss_name) 
        for img in os.listdir(path): 
            name, _ = os.path.splitext(img)
            img_path = [path +"/"+name+".jpeg"]
            filenames = filenames+img_path
            img_array = io.imread(os.path.join(path,img))
            H = img_array.shape[0]
            W = img_array.shape[1]
            class_num = np.zeros((H,W))
            for i in range(H):
                for j in range(W):
                    if img_array.item(i,j) == 0:
                        class_num[i,j]=1.0
                    else:
                        class_num[i,j]=0.7
            class_num = class_num.reshape(H,W)
            file_path.append([img_path[0],class_num])     
                   
     
print("Number of input patches:",len(file_path))
data_address=[]
data_label=[]

for address,label in file_path:
    data_address.append(address)
    data_label.append(label)
    
f=imread(data_address[0])
plt.imshow(f)
plt.show()
print(data_label[0])
S=data_label[0]
print(S[100:200])
print("data_label.shape",S.shape)
```

```python

train_data, validation_data, train_label, validation_label = train_test_split(data_address, data_label, test_size=0.3)
test_data, validation_data, test_label, validation_label = train_test_split(validation_data, validation_label, test_size=0.5)


def parse_function_T(train_data, train_label):
   
    image_string = tf.io.read_file(train_data)
    image = tf.image.decode_jpeg(image_string, channels=1)
    image = tf.image.convert_image_dtype(image, tf.float32)
    label = train_label
    
    return image, label


def parse_function_V(validation_data, validation_label):
   
    image_string = tf.io.read_file(validation_data) 
    image = tf.image.decode_jpeg(image_string, channels=1)
    image = tf.image.convert_image_dtype(image, tf.float32)
    label = validation_label
    
    return image, label


def train_preprocess(image, label):
    
    image = image[...,1]
    image = image[..., tf.newaxis]
    
    return image, label


def show(image,label):
    
    plt.figure()
    plt.imshow(image)
    plt.title(label)
    plt.axis('off')

#########Train data########

train_dataset = tf.data.Dataset.from_tensor_slices((train_data, train_label))
train_dataset = train_dataset.shuffle(len(train_data))
train_dataset = train_dataset.map(parse_function_T)
#train_dataset = train_dataset.map(train_preprocess)
train_dataset = train_dataset.batch(32)
#dataset = dataset.prefetch(1)

####Validation data#######

validation_dataset = tf.data.Dataset.from_tensor_slices((validation_data, validation_label))
validation_dataset = validation_dataset.shuffle(len(validation_data))
validation_dataset = validation_dataset.map(parse_function_V)
#validation_dataset = validation_dataset.map(train_preprocess)
validation_dataset = validation_dataset.batch(32)
#dataset = dataset.prefetch(1)    

for image,label in validation_dataset.take(1):
    
    print(image.shape)
    #show(image,label.numpy())
     
for image,label in train_dataset.take(1):

    print(image.shape)
    #show(image,label.numpy())
    
```

# Training Network

```python
inpt = layers.Input((200, 200, 1))
conv1 = layers.Conv2D(64,(3,3),activation="relu",padding="same")(inpt)
pooling1 = layers.MaxPooling2D(pool_size=(2, 2))(conv1)

conv2 = layers.Conv2D(64,(3,3),activation="relu",padding="same")(pooling1)
pooling2 = layers.MaxPooling2D(pool_size=(2, 2))(conv2)

conv3 = layers.Conv2D(64,(3,3),activation="relu",padding="same")(pooling2)
pooling3 = layers.MaxPooling2D(pool_size=(2, 2))(conv3)

deconv1 = layers.Conv2DTranspose(64,(3,3),strides=2,activation="relu",padding="same")(pooling3)
deconv2 = layers.Conv2DTranspose(64,(3,3),strides=2,activation="relu",padding="same")(deconv1)
deconv3 = layers.Conv2DTranspose(64,(3,3),strides=2,activation="relu",padding="same")(deconv2)
H = conv1.shape[1]
W = conv1.shape[2]
deconv3_new = tf.image.resize( deconv3, [H,W])
deconv3_new= layers.concatenate([deconv3_new, conv1])
#deconv3= layers.concatenate([deconv3, conv1])
out = layers.Conv2D(1,1, activation="relu")(deconv3_new)
model = models.Model(inputs=inpt, outputs=out)

print("out",out.shape)
print("inpt",inpt.shape)
model.compile(loss="mse",optimizer="Adam")
model.summary()

filepath = "ModelWeights-{epoch:02d}-{loss:.4f}.hdf5"
checkpoint = ModelCheckpoint(filepath, save_best_only=True, monitor="loss")
history = model.fit(train_dataset, callbacks=[checkpoint], epochs=3,batch_size=32, validation_data=validation_dataset)

model.save('my_model_Nm1.h5')  
del model  

plt.plot(history.history['loss'])
plt.title('loss')
plt.ylabel('loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Val'], loc='upper left')
plt.show()
```
